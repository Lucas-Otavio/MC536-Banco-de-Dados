```python
import requests
import json
```

## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"

concept = "Tent of Miracles (novel)"
format = "jsod"

concept_ = concept.replace(" ", "_")

url = url_template.replace("{concept}", concept_)\
                  .replace("{format}", format)

data = requests.get(url)
js = data.json()
print(json.dumps(js, indent=2))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres

```python
#Recupera endereço do recurso do autor
author_resource = js["d"]["results"][0]["http://dbpedia.org/property/author"]["__deferred"]["uri"]

#Encontra index do ultimo char que precede o nome do autor
author_index = author_resource.rfind("/")

#Recorta a string, restando apenas o autor, e substitui '_' por ' '
author = author_resource[author_index + 1:].replace("_", " ")

#Imprime quem é o autor do conceito
print(f"Autor de {concept} é {author}")
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
#Faz requisição para encontrar o código do termo
url_template = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label={concept}&match={match}&limit={limit}"

concept = "Aspirin"
match = "exact"
limit = "10"

concept = concept.replace(" ", "%20")

url = url_template.replace("{concept}", concept)\
                  .replace("{match}", match)\
                  .replace("{limit}", limit)

data = requests.get(url)
json_aspirin = data.json()
```

```python
#Recupera endereço do recurso do termo
code_resource = json_aspirin[0]["resource"]

#Encontra index do ultimo char que precede o termo
code_index = code_resource.rfind("/")

#Recorta a string, restando apenas o termo
code = code_resource[author_index:]

#Executa a chamada
url = f"https://id.nlm.nih.gov/mesh/lookup/details?descriptor={code}"
data = requests.get(url)

#Acessa o json
json_mid = data.json()

#Recupera a lista de sinônimos
terms_list = json_mid["terms"]
terms = [element["label"] for element in terms_list]

#Imprime os sinônimos
print(f"São termos usados para \"{concept}\":")
for i in range(len(terms)):
    print(f"%02d. {terms[i]}" %(i+1))
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
#Chamada ao Open Street Map para Query
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Pitt Rivers Museum"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
museum_ref = data.json()
print(json.dumps(museum_ref, indent=2))
```

```python
#Chamada ao Open Street Map para Elementos
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"
element_type = "way" # types: node, way, or relation
id = str(museum_ref[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
museum_way = data.json()
print(json.dumps(museum_way, indent=3))
```

```python
#Chamada ao DBPedia
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = museum_way["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "json"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
museum_dbpedia = data.json()
print(json.dumps(museum_dbpedia, indent=3))
```
